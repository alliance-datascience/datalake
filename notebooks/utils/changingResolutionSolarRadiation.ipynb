```python
import xarray as xr
import numpy as np
```

```python
localNCPath = 'data/Solar-Radiation-Flux_C3S-glob-agric_AgERA5_19800101_final-v1.0.nc'
```

```python
geoData = xr.open_dataset(localNCPath)
```

```python
standardizedResolution =   xr.Dataset(
    {
        "lat": (["lat"], np.arange(-90, 90, 0.05)),
        "lon": (["lon"], np.arange(-180, 180, 0.05)),
    }
)

lat = np.arange(-90, 90, 0.05)
lon = np.arange(-180, 180, 0.05)
```

```python
standardizedResolution
```

```python
dataInterpolated = geoData.interp(standardizedResolution,method='cubic')
```

```python
dataInterpolated = geoData.interp(lat=lat
               ,lon= lon
              ,method='cubic')
```
