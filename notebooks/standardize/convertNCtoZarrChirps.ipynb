```python
compressor = zarr.Blosc(cname='lz4', clevel= 1, shuffle=False)
from numcodecs import blosc
blosc.set_nthreads(8) 
encoding = {vname: {'compressor': compressor,'chunks': (1,1,2000,7200)} for vname in allFiles.data_vars}
# encoding = {vname: {'chunks': (1,1,2000,7200)} for vname in allFiles.data_vars}
#encoding = {vname: {'compressor': None,'chunks': (1,1,2000,7200)} for vname in allFiles.data_vars}
```

```python
s3 = s3fs.S3FileSystem()
store = s3fs.S3Map(root="s3://climate-action-datalake/zone=raw/source=chirps/variable=precipitation", s3=s3, check=False)
```

```python
with ProgressBar():
    result = allFiles.to_zarr('data/precipitation.zarr', encoding=encoding, consolidated=True,compute=False)
```
